```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = 'Data sold to the grid.xlsx'
df = pd.read_excel(file_path)


print(df.head())

# Set the style for the plots
sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))
sns.lineplot(x='month_name', y='Generator Export to Grid (MWh)', data=df, marker='o')
plt.title("Monthly Generator Export to Grid (MWh)", fontsize=16)
plt.ylabel("Generator Export to Grid (MWh)", fontsize=14)
plt.xlabel("Month", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='month_name', y='Payment from AESO (Thousands $)', data=df, marker='x')
plt.title("Monthly Payment from AESO (Thousands $)", fontsize=16)
plt.ylabel("Payment from AESO (Thousands $)", fontsize=14)
plt.xlabel("Month", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
