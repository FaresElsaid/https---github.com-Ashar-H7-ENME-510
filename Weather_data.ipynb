```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('weatherstats_calgary_daily (1).csv')

df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

monthly_avg_temp_over_years = df.groupby(['year', 'month'])['avg_temperature'].mean().reset_index()

plt.figure(figsize=(14, 8))
sns.lineplot(data=monthly_avg_temp_over_years, x='month', y='avg_temperature', hue='year', marker='o', palette='tab10')
plt.title('Monthly Average Temperature Over the Years in Calgary', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Temperature (°C)', fontsize=14)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
sns.despine()
plt.tight_layout()
plt.show()
```
