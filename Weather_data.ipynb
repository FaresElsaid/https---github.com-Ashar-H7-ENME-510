```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('weatherstats_calgary_daily (1).csv')

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract the year and month from the 'date' column
df['year_month'] = df['date'].dt.to_period('M').astype(str)

# Calculate the average temperature for each month over the years
monthly_avg_temp_over_years = df.groupby('year_month')['avg_temperature'].mean().reset_index()

# Plot the monthly data over the years using seaborn
plt.figure(figsize=(14, 8))
sns.lineplot(data=monthly_avg_temp_over_years, x='year_month', y='avg_temperature', marker='o', palette='tab10')
plt.title('Monthly Average Temperature Over the Years in Calgary', fontsize=16)
plt.xlabel('Date (Year-Month)', fontsize=14)
plt.ylabel('Average Temperature (C)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
sns.despine()
plt.tight_layout()
plt.show()
```
